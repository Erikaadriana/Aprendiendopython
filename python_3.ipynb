```python
#Siempre lo primero que deb hacer es importar pandas y numpy
import pandas as pd
import numpy as np
```

```python
#ahora subimos el archivo desde google drive , ya no tenemos que logearnos de nuevo
df_estudiantes=pd.read_csv("/content/drive/MyDrive/StudentsPerformance.csv")
df_estudiantes
```

```python
#simagomos con la creacion de columnas usando el metodo assign o el metodo insert
#si quiero agregar varias columnas uso assing y tambien si quiero sobreescribir una columna
```

```python
#creamos notas de fisica y filosofia aleatorias y las guardamos en variables
filosofia_score=np.random.randint(1, 100, size=1000)
Fisica_score=np.random.randint(1,100, size=1000)
```

```python
#ahora le pondre un indice a cada fila para saber donde va en el dataframe
filosofia_nota=pd.Series(filosofia_score, index=np.arange(0,1000))
fisica_nota=pd.Series(Fisica_score, index=np.arange(0,1000))
```

```python
#Ahora lo que voy a hacer es poner las nuevas columnas a mi dataframe usando asing
df_estudiantes.assign(filosofia_score=filosofia_nota,Fisica_score=fisica_nota)
```

```python
df_estudiantes=df_estudiantes.assign(filosofia_score=filosofia_nota,Fisica_score=fisica_nota)
```

```python
df_estudiantes
```

```python
#ahora vamos a usar el metodo insert , el metodo insert es diferente porque me permite especificar donde voy a colocar la columna
#al usar el metodo insert debemos tener cuidado por que no me devuelve copias si no que el cambio queda hecho de una vez
df_estudiantes.insert(1, "ED_SCORE", fisica_nota)
df_estudiantes
```
