```python
#tambien elimina columnas y filas con doferentes metodos
import pandas as pd
import numpy as np
```

```python
#traemos el archivo
df_jugadores=pd.read_csv("/content/players_20 (1).csv")
df_jugadores.head(4)
```

```python
#ahora fijo el index usando el shortname
df_jugadores=df_jugadores.set_index("short_name")
df_jugadores=df_jugadores[["player_url","long_name","age","dob","height_cm","weight_kg","nationality","club"]]
df_jugadores.head(3)
```

```python
#vamos a eliminar regostros usando drop
df_jugadores.drop("L. Messi", axis=0)
```

```python
#ahora vamos a eliminar teniendo en cuenta el indes
df_jugadores.drop(index=["Neymar Jr"])
```

```python
#ahora vams a eliminar dos o mas jugadores
df_jugadores=df_jugadores.drop(["L. Messi","Neymar Jr"],axis=0)
```

```python
df_jugadores=df_jugadores.drop(["E. Hazard","K. De Bruyne"],axis=0)
df_jugadores
```

```python
#para eliminar una columna
df_jugadores.drop("player_url",axis=1)#cuando axis es igual a uno quiere decir que elimino toda la columna
```

```python
#otro metodo para eliminar las columnas sin usar el acis es el siguiente
df_jugadores.drop(columns=["nationality"])
```

```python
#para eliminar la ultima columna si no la sabemos lo hacemosde la siguiente manera
df_jugadores.drop(df_jugadores.columns[[-1]], axis=1)
```
