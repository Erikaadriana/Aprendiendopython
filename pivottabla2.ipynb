```python
#este metodo es unpoco mas robusto nos permite hace operaciones y no solo mostrar
import pandas as pd
import numpy as np
df_nuevo=pd.read_excel("/content/supermarket_sales.xlsx")
```

```python
df_nuevo.head(2)
```

```python
#vamos a hacer una tabla pivote usando los registros
#observaremos quien compra mas popr genero
#vamos a usar dentro del metodopivot table un argumneto que es raro que se llama aggfunc=que me ayuda a determinar la operacion
#vamos primero a contar
df_nuevo.pivot_table(index="Gender", aggfunc="count")
```

```python
#aohra lo qu evamos a hacer es sumar en ccada columna por genero sin embrgo recuerden que solo puedo sumar nuemros
df_nuevo.pivot_table(
    index="Gender",
    values=["Quantity","Rating","Tax 5%","Total","Unit price","cogs" ,"gross income","gross margin percentage"],
    aggfunc="sum"
    )
```

```python
df_nuevo.pivot_table(
    index="Gender",
    values=["Quantity","Total"],
    aggfunc="sum"
    )
```

```python
#agreguemos mas datos al ejercicio
df_nuevo.pivot_table(
    index="Gender",
    columns="Product line",
    values=("Total"),
    aggfunc="sum"
)
```
