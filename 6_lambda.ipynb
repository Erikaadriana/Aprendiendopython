```python
#en este caso mexclaremos la funcin lambida con el metidi apply
import pandas as pd
import numpy as np
```

```python
df_jugadores=pd.read_csv("/content/players_20 (1).csv")
df_jugadores.head(4)
```

```python
df_jugadores.set_index("short_name" , inplace=True)
#seleccciono las columnas que voy a usar
df_jugadores=df_jugadores[["player_url","long_name","age","dob","height_cm","weight_kg","nationality","club"]]
```

```python
#ahora vamos a ver como se hacen las funciones usando lambda
#primero se asigna el nombre luego asignamos el valor luego los parametros lueg dos puntos y por ultimo el output
sumar_2=lambda a,b:a+b
```

```python
sumar_2(39,39)
```

```python
modulo_2=lambda a,b:a%b
modulo_2(25,2)
```

```python
#ahora usemos apply con lambda
#vamos a pasar los centimetros a metros en la informacion del jugador
#en este caso si usamos apply no necesitamos poner el nombre a la funcion
df_jugadores["height_cm"].apply(lambda x:x/100)
```

```python
df_jugadores["age"].apply(lambda x:x*12)
```

```python
#si quisiera enredarme menos para convertir los cm a metros
df_jugadores["height_cm"]/100
```

```python
df_jugadores["age"]*12
```

```python
#tambien podemos pasar los nombts de los jugadores a mayusculas
df_jugadores["long_name"].apply(lambda x:x.upper)
```

```python
df_jugadores["long_name"].str.upper()
```

```python
#tambien puedo usar lambda para extraer partes de fechas
#primero debemos garantizar que la columna fecha este en modo date
df_jugadores["dob"]=pd.to_datetime(df_jugadores["dob"])
```

```python
#ahora extraemos los años solo con la funcion
df_jugadores["dob"].apply(lambda x:x.month)
```

```python
df_jugadores["dob"].apply(lambda x:x.month_name())
```

```python
#peso sobre altura al cuadrado
df_jugadores.apply(lambda x:x["weight_kg"]/((x["height_cm"]/100)**2),axis=1)
```
