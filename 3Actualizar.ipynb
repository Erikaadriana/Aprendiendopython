```python
# lo que haremos ahora es modificar valores luego de qu elos extraemos
#loprimero es imkportar pantas y numpy
import pandas as pd
import numpy as np
```

```python
#luego traemos el archivo
df_jugadores=pd.read_csv("/content/players_20 (1).csv")
```

```python
#ahora fijo el index usando el shortname
df_jugadores=df_jugadores.set_index("short_name")
```

```python
df_jugadores=df_jugadores[["player_url","long_name","age","dob","height_cm","weight_kg","nationality","club"]]
df_jugadores.head(3)
```

```python
#ahora vamos a actualizar la estatura de mesi
df_jugadores.loc["L. Messi", "weight_kg"	]
```

```python
#asi se modifica un dato con loc
df_jugadores.loc["L. Messi", "weight_kg"]=89
```

```python
#muestra los datos
df_jugadores.loc["L. Messi", :	]
```

```python
df_jugadores.loc["Cristiano Ronaldo", "height_cm"	]
df_jugadores.loc["Cristiano Ronaldo", "height_cm"]=190
df_jugadores.loc["Cristiano Ronaldo", :	]
```

```python
#actualizar el valor de toda una columna
df_jugadores.loc[:,"height_cm"]=190
df_jugadores
```

```python
#como modifico el ultimo registro
df_jugadores.iloc[-1, :]
```

```python
#de esta manera puedo poner datos nulos en el dataframe
df_jugadores.iloc[-1, :]=np.nan
df_jugadores.iloc[-1,:]
```

```python
#ahora veamos como puedo actualizar multiples valores
#por ejemplo voy a actualizar a mesi y a mi comandante
df_jugadores.iloc[[0,1], 4]=100
```

```python
df_jugadores.head(4)
```

```python
#vamos  a actualizar valores cuando se cumple ocn una condicion
#primero definimos las columnas  que queremosmostrar e n una variable
columnas=["age","height_cm","weight_kg","nationality","club"]
df_jugadores.loc[df_jugadores["height_cm"]<180,columnas]=0
```
